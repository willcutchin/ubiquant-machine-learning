```python
# Import libraries
import pandas as pd
import numpy as np
import csv
```

```python
# Load the original "Training Data Set"
full_path = "~/Desktop/School/C365_data_analytics_and_mining/final_project/ubiquant-machine-learning/data/training_data/original_training_set/train.csv"

df = pd.read_csv(full_path)
```

```python
# Check Load
df.head()
```

```python
# Creating 15 Random Sample Files
sample_count = 15
sample_size = 50000
counter = 0
for i in range(0, sample_count):
    sample = df.sample(n=sample_size)
    counter += 1
    sample.to_csv(r'~/Desktop/School/C365_data_analytics_and_mining/final_project/ubiquant-machine-learning/data/training_data/random_sample_set/samples/random_training_sample_' + str(counter) +".csv")
    print("sample" + str(counter) + "saved")
```

```python
description_path = "~/Desktop/School/C365_data_analytics_and_mining/final_project/ubiquant-machine-learning/data/training_data/random_sample_set/descriptions/"
```

```python
# Take the description of the Training file and save it to csv
df.describe().to_csv(description_path + "train_data_description.csv")
```

```python
# Take the description of each Sample and save it as a csv
sample_path = "~/Desktop/School/C365_data_analytics_and_mining/final_project/ubiquant-machine-learning/data/training_data/random_sample_set/samples/"
sample_count = 15
counter = 0
for i in range(0, sample_count):
    counter += 1
    sample = pd.read_csv(sample_path + "random_training_sample_" + str(counter) +".csv")
    sample.describe().to_csv(description_path + "random_training_sample_description_" + str(counter) + ".csv")
    print("sample description saved: " + str(counter))
```
